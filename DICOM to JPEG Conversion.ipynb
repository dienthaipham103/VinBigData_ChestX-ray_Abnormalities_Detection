### Convert dicom into images

```
def convert(input_dir, output_dir):
  count = 0
  for filename in os.listdir(input_dir):
    dicom_path = os.path.join(input_dir, filename)
    saved_path = os.path.join(output_dir, filename[:-5] + 'png')

    ds = pydicom.read_file(dicom_path) # read dicom image

    img = ds.pixel_array.astype(float) # get image array
    # Rescaling grey scale between 0-255
    img_2d_scaled = (np.maximum(img, 0) / img.max()) * 255.0
    # Convert to uint
    img_2d_scaled = np.uint8(img_2d_scaled)

    print(img.shape)
    cv2.imwrite(saved_path, img_2d_scaled) # write png image

    count += 1
    print('count: ', count)
```

```
input_dir = root_path + 'data/train'
output_dir = root_path + 'data/train_images'

convert(input_dir, output_dir)
```

### count files

```
def count_files(folder):
  count = 0
  for _ in os.listdir(folder):
    count += 1
  print('the number of images in %s : %d' % (folder, count))
  return count

count_files(root_path + 'data/test_images')
```
